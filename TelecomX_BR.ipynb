#ðŸ”§ TransformaÃ§Ã£o

```python
# tranformando dados em um dataframe
import json

with open('data/TelecomX_Data.json', mode='r') as file:
    dados = json.load(file)
dados
```

```python
# normalizando os dados
dados_normalizados = pd.json_normalize(dados)
dados_normalizados.head()
```

```python
# analizando os dados
dados_normalizados.info()
```

```python
# verificando valores nulos
dados_normalizados.isna().sum()
```

```python
# verificando valores unicos
dados_normalizados.nunique()
```

```python
# verificando os valores de cada coluna
print("Valores Ãºnicos em cada coluna:")
for col in dados_normalizados.columns:
  print(f"{col}: {dados_normalizados[col].unique()}")
  print(f"{col}: {dados_normalizados[col].nunique()}")
  print("-" * 30)
```

```python
# Verificando valores vazios na coluna 'account.Charges.Total'
dados_normalizados[dados_normalizados['account.Charges.Total'] == ' ']
```

```python
dados_normalizados[dados_normalizados['account.Charges.Total'] == ' ']
```

```python
# Pegando os Ã­ndices dos valores vazios na coluna 'account.Charges.Total'
index = dados_normalizados.loc[dados_normalizados['account.Charges.Total'] == ' '].index
index
```

```python
# Removendo os valores vazios na coluna 'account.Charges.Total' e substituindo por 0
dados_normalizados['account.Charges.Total'] = dados_normalizados['account.Charges.Total'].replace(' ', 0)
```

```python
# Convertendo a coluna 'account.Charges.Total' para o tipo numÃ©rico float
dados_normalizados['account.Charges.Total'] = dados_normalizados['account.Charges.Total'].astype(np.float64)
```

```python
# Convertendo a coluna 'account.Charges.Monthly' para o tipo numÃ©rico float
dados_normalizados['account.Charges.Monthly'] = dados_normalizados['account.Charges.Monthly'].astype(np.float64)
```

```python
dados_normalizados.info()
```

```python
dados_normalizados.query("Churn == ''")
```

```python
# Filtrando os dados para remover linhas onde a coluna 'Churn' estÃ¡ vazia
# e mantendo apenas as linhas onde 'Churn' nÃ£o Ã© vazio
dados_churn = dados_normalizados[dados_normalizados['Churn'] != ''].copy()
dados_churn
```

```python
dados_churn.info()
```

```python
# Salvando os dados limpos em um novo arquivo CSV
dados_churn.to_csv('data/TelecomX_dados_limpos.csv', index=False)
```
