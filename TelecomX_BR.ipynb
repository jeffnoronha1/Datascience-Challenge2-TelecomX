#ðŸ”§ TransformaÃ§Ã£o

```python
# tranformando dados em um dataframe
import json

with open('data/TelecomX_Data.json', mode='r') as file:
    dados = json.load(file)
# dados
```

```python
# normalizando os dados
dados_normalizados = pd.json_normalize(dados)
dados_normalizados.head()
```

```python
# analizando os dados
dados_normalizados.info()
```

```python
# verificando valores nulos
dados_normalizados.isna().sum()
```

```python
# verificando valores unicos
dados_normalizados.nunique()
```

```python
# verificando os valores de cada coluna
print("Valores Ãºnicos em cada coluna:")
for col in dados_normalizados.columns:
  print(f"{col}: {dados_normalizados[col].unique()}")
  print(f"{col}: {dados_normalizados[col].nunique()}")
  print("-" * 30)
```

```python
# Verificando valores vazios na coluna 'account.Charges.Total'
dados_normalizados[dados_normalizados['account.Charges.Total'] == ' ']
```

```python
dados_normalizados[dados_normalizados['account.Charges.Total'] == ' ']
```

```python
# Pegando os Ã­ndices dos valores vazios na coluna 'account.Charges.Total'
index = dados_normalizados.loc[dados_normalizados['account.Charges.Total'] == ' '].index
index
```

```python
# Removendo os valores vazios na coluna 'account.Charges.Total' e substituindo por 0
dados_normalizados['account.Charges.Total'] = dados_normalizados['account.Charges.Total'].replace(' ', 0)
```

```python
# Convertendo a coluna 'account.Charges.Total' para o tipo numÃ©rico float
dados_normalizados['account.Charges.Total'] = dados_normalizados['account.Charges.Total'].astype(np.float64)
```

```python
# Convertendo a coluna 'account.Charges.Monthly' para o tipo numÃ©rico float
dados_normalizados['account.Charges.Monthly'] = dados_normalizados['account.Charges.Monthly'].astype(np.float64)
```

```python
dados_normalizados['daily.accounts'] = dados_normalizados['account.Charges.Monthly'] / 30
dados_normalizados['daily.accounts'] = dados_normalizados['daily.accounts'].round(2)
```

```python
dados_normalizados.info()
```

```python
dados_normalizados.query("Churn == ''")
```

```python
# Filtrando os dados para remover linhas onde a coluna 'Churn' estÃ¡ vazia
# e mantendo apenas as linhas onde 'Churn' nÃ£o Ã© vazio
dados_churn = dados_normalizados[dados_normalizados['Churn'] != ''].copy()
dados_churn
```

```python
dados_churn.info()
```

```python
# Salvando os dados limpos em um novo arquivo CSV
dados_churn.to_csv('data/TelecomX_dados_limpos.csv', index=False)
```

#ðŸ“Š Carga e anÃ¡lise

```python
# Carregando os dados limpos do arquivo CSV
dados = pd.read_csv('data/TelecomX_dados_limpos.csv')
dados.head(5)
```

```python
dados.shape
```

```python
# anÃ¡lise descritiva dos dados
dados.describe()
```

```python
churn_rate = dados['Churn'].value_counts(normalize=True) * 100
print("Taxa de Churn (%): {}".format(churn_rate))
```

```python
# Visualizando a taxa de churn
churn_rate.plot(
    kind='pie',
    colors=['#5cb85c', '#d9534f'],
    autopct='%.1f%%',
    startangle=90,
    labels=['NÃ£o Churn', 'Churn'],
    legend=False)
plt.title('Taxa de Churn')
plt.ylabel('')
plt.tight_layout()
plt.savefig('graficos/churn_rate.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Contagem total de clientes
total_customers = dados['customerID'].nunique()
print("Total de clientes: {}".format(total_customers))
```

```python
# Contagem total de clientes que cancelaram
total_churned = dados[dados['Churn'] == 'Yes']['customerID'].nunique()
print("Total de clientes que cancelaram: {}".format(total_churned))
```

```python
# Contagem total de clientes nÃ£o cancelados
total_not_churned = dados[dados['Churn'] == 'No']['customerID'].nunique()
print("Total de clientes que nÃ£o cancelaram: {}".format(total_not_churned))
```

```python
# AnÃ¡lise de churn por gÃªnero
churn_genero = dados.groupby('customer.gender')['Churn'].value_counts(normalize=True).unstack() * 100
churn_genero
```

```python
# Visualizando a taxa de churn por gÃªnero
ax = churn_genero.plot(kind='bar', color=['#5cb85c', '#d9534f'])
plt.title('Taxa de Churn por GÃªnero')
plt.xlabel('GÃªnero')
plt.ylabel('Porcentagem (%)')
plt.legend(title='Churn', labels=['NÃ£o Churn', 'Churn'])
plt.xticks(ticks=[0, 1], labels=['Masculino', 'Feminino'], rotation=0)
plt.xticks(rotation=0)

for p in ax.patches:
    ax.annotate(f"{p.get_height():.1f}%", 
                (p.get_x() + p.get_width() / 2, p.get_height() + 1),
                ha='center', va='bottom', fontsize=10, color='black')

plt.ylim(0, 100)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.savefig('graficos/churn_genero.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# AnÃ¡lise de churn por idade
churn_idade = dados.groupby('customer.SeniorCitizen')['Churn'].value_counts(normalize=True).unstack() * 100
churn_idade
```

```python
# Visualizando a taxa de churn por idade
ax = churn_idade.plot(kind='bar', color=['#5cb85c', '#d9534f'])
plt.title('Taxa de Churn por Idade')
plt.xlabel('Idade')
plt.ylabel('Porcentagem (%)')
plt.xticks(ticks=[0, 1], labels=['NÃ£o Idoso', 'Idoso'], rotation=0)
for p in ax.patches:
    ax.annotate(f"{p.get_height():.1f}%", 
                (p.get_x() + p.get_width() / 2, p.get_height() + 1),
                ha='center', va='bottom', fontsize=10, color='black')
plt.legend(title='Churn', labels=['NÃ£o Churn', 'Churn'], loc='upper right')
plt.ylim(0, 100)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.savefig('graficos/churn_idade.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# AnÃ¡lise de churn por tipo de contrato
churn_tipo_contrato = dados.groupby('account.Contract')['Churn'].value_counts(normalize=True).unstack() * 100
churn_tipo_contrato
```

```python
# Visualizando a taxa de churn por gÃªnero
ax = churn_tipo_contrato.plot(kind='bar', color=['#5cb85c', '#d9534f'])
plt.title('Taxa de Churn por Tipo de Contrato')
plt.xlabel('Tipo de Contrato')
plt.ylabel('Porcentagem (%)')
plt.xticks(
    ticks=range(len(churn_tipo_contrato.index)),
    labels=['MÃªs a MÃªs', 'Um Ano', 'Dois Anos'],
    rotation=0
)
for p in ax.patches:
    ax.annotate(f"{p.get_height():.1f}%", 
                (p.get_x() + p.get_width() / 2, p.get_height() + 1),
                ha='center', va='bottom', fontsize=10, color='black')
plt.legend(title='Churn', labels=['NÃ£o Churn', 'Churn'], loc='upper right')
plt.ylim(0, 100)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.savefig('graficos/churn_tipo_contrato.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# AnÃ¡lise de churn por tipo de pagamento
mapeamento = {
    'Electronic check': 'Cheque EletrÃ´nico',
    'Mailed check': 'Cheque Enviado',
    'Bank transfer (automatic)': 'TransferÃªncia BancÃ¡ria (AutomÃ¡tica)',
    'Credit card (automatic)': 'CartÃ£o de CrÃ©dito (AutomÃ¡tico)'
}

churn_tipo_pagamento = dados.groupby('account.PaymentMethod')['Churn'].value_counts(normalize=True).unstack() * 100
churn_tipo_pagamento.rename(index=mapeamento, inplace=True)
churn_tipo_pagamento.sort_values(by='Yes', ascending=True, inplace=True)
churn_tipo_pagamento = churn_tipo_pagamento.rename(
    columns={'No': 'NÃ£o Churn', 'Yes': 'Churn'}
)
churn_tipo_pagamento
```

```python
# Contagem total de churn por tipo de pagamento
total_churn_tipo_pagamento = dados.groupby('account.PaymentMethod')['Churn'].value_counts().unstack()
total_churn_tipo_pagamento
```

```python
ax = churn_tipo_pagamento.plot(kind='barh', color=['#5cb85c', '#d9534f'])
plt.title('Taxa de Churn por Tipo de Pagamento')
plt.xlabel('Porcentagem (%)')
plt.ylabel('Tipo de Pagamento')

for p in ax.patches:
    width = p.get_width()
    y_center = p.get_y() + p.get_height() / 2
    ax.annotate(f"{width:.1f}%",
                (width + 1, y_center),
                ha='left', va='center', fontsize=9, color='black') 

plt.legend(title='Churn', labels=['NÃ£o Churn', 'Churn'], loc='upper right') # Ajustado a posiÃ§Ã£o da legenda
plt.xlim(0, 100)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('graficos/churn_tipo_pagamento.png', dpi=300, bbox_inches='tight') # Salva com um novo nome para diferenciar
plt.show()
```

```python
# AnÃ¡lise de churn por tenure pegando os primeiros 12 meses
churn_tenure = dados.groupby('customer.tenure')['Churn'].value_counts(normalize=True).unstack() * 100
churn_tenure = churn_tenure[churn_tenure.index <= 12]  
churn_tenure
```

```python
# Visualizando a taxa de churn por tempo de permanÃªncia (tenure)
ax = churn_tenure.plot(kind='bar', color=['#5cb85c', '#d9534f'])
plt.title('Taxa de Churn por Tempo de PermanÃªncia (12 meses)')
plt.xlabel('Tempo de PermanÃªncia (Meses)')
plt.ylabel('Porcentagem (%)')
plt.xticks(rotation=0)
for p in ax.patches:
    ax.annotate(f"{p.get_height():.1f}%", 
                (p.get_x() + p.get_width() / 2, p.get_height() + 1),
                ha='center', va='bottom', fontsize=10, color='black')
plt.legend(title='Churn', labels=['NÃ£o Churn', 'Churn'], loc='upper right')
plt.ylim(0, 100)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.savefig('graficos/churn_tenure.png', dpi=300, bbox_inches='tight')
plt.show()
```

## IntroduÃ§Ã£o

O objetivo deste trabalho foi analisar o fenÃ´meno de evasÃ£o de clientes (churn) na base de dados da TelecomX. O churn representa o cancelamento de contratos por parte dos clientes, sendo um dos principais desafios para empresas do setor de telecomunicaÃ§Ãµes. A anÃ¡lise visa identificar padrÃµes e fatores associados ao churn, fornecendo subsÃ­dios para estratÃ©gias de retenÃ§Ã£o.

---

## Limpeza e Tratamento de Dados

- **ImportaÃ§Ã£o:** Os dados foram extraÃ­dos do arquivo `TelecomX_Data.json`.

- **Tratamento de valores ausentes:** Valores vazios na coluna `account.Charges.Total` foram substituÃ­dos por 0 e convertidos para tipo numÃ©rico.
